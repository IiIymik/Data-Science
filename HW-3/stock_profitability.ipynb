```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = {
      "Момент наблюдения": [1,2,3,4,5,6],
      "Доходность A, %": [25, -10, 10, 5, 35, 13],
      "Доходность B, %": [0, 15, -5, 5, 20, 25],
      "Доходность C, %": [10, 25, -15, -5, -5, 15]
      }
tabl = pd.DataFrame(data)
tabl
```

```python
sns.lmplot(x="Доходность A, %", y="Доходность B, %", data=tabl)
```

```python
sns.lmplot(x="Доходность A, %", y="Доходность C, %", data=tabl)
```

```python
sns.lmplot(x="Доходность B, %", y="Доходность C, %", data=tabl)
```

Вопрос: Выберите пару акций которые наименее зависимы в доходностях от друг друга?

Ответ: Судя по графикам регрессии, худшая зависимость между акциями A / B.

```python
tabl.corr()
```
