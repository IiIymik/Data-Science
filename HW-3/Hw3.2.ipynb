```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

```python
coordinates = []
triangle = [[50, 50], [500, 950], [950, 50]]
coordinates.extend(triangle)
x, y = random.randint(0,1000), random.randint(0,1000)
coordinates.append([x, y])
coordinates
for i in range(1000000):
  A = random.randint(0,2)
  x, y = 0.5*(x + triangle[A][0]), 0.5*(y + triangle[A][1])
  coordinates.append([x, y])
coord_X = [i[0] for i in coordinates]
coord_Y = [i[1] for i in coordinates]
tabl = pd.DataFrame({'X': coord_X, 'Y': coord_Y})
tabl
```

```python
from matplotlib import figure
plt.style.use('dark_background')
plt.figure(figsize=(30, 14))
plt.scatter(coord_X, coord_Y, c='aqua', s=1)
plt.show()
```
