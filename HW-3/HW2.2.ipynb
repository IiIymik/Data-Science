```python
import pandas as pd
import numpy as np
```

```python
URL = 'https://drive.google.com/file/d/1zNVdFzREtUUlBJ54xvYQ3JlD4HWh8eNq/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id='+URL.split('/')[-2]
tbl = pd.read_csv(path)
```

```python
tbl.dtypes
```

```python
tbl.isnull().sum()
```

```python
drop_list = [col for col in tbl.columns if tbl[col].isnull().sum() != 0 and col != 'Язык.программирования']
tbl.drop(drop_list, axis=1, inplace=True)
tbl.isnull().sum()
```

```python
tbl.dropna(inplace=True)
tbl.isnull().sum()
```

```python
tbl.shape
```

```python
python_data = tbl[tbl['Язык.программирования'] == 'Python']
python_data
```

```python
python_data.shape
```

```python
position = python_data.groupby('Должность')
```

```python
position['salary'].agg(['min', 'max'])
```

```python
def fill_avg_salary(row):
  return (row['min'] + row['max']) / 2
```

```python
position['salary'].agg(['min', 'max']).apply(fill_avg_salary, axis=1)
```

```python
position_avg = position['salary'].agg(['min', 'max']).apply(fill_avg_salary, axis=1)
position_avg.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gender = tbl.groupby('Пол').count()
plt.bar(list(gender.index.values), gender['N'], color=['r', 'b'])
plt.title('Співвідношення кількості жінок та чоловіків')
plt.show()
```

```python
plt.barh(list(position_avg.index.values), list(position_avg), color='g')
plt.title('Salaries')
plt.xticks(rotation = 90)
plt.show()
```

```python
age = tbl.groupby('Возраст')['salary'].mean()
plt.plot(list(age.index.values), list(age), color='r')
plt.xlabel('Вік', fontsize='small')
plt.ylabel('USD', fontsize='small')
plt.grid()
plt.title('Зарплатня від віку', fontsize=15)
plt.show()
```
